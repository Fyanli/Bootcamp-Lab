```python
#Import all the libraries that are necessary.
#Open the Employee.csv file and name your dataset employee
import pandas as pd
import numpy as np
employee = pd.read_csv('Employee.csv')
employee.head()
```

```python
#What's the mean salary by department?

# Method 1
employee.groupby(['Department']).agg({'Salary':avg})
```

```python
# Method 2
employee.pivot_table(index= ['Department'], values = ['Salary'], aggfunc = {'Salary':'mean'})
```

```python
# We want to display the number of years of work experience and the salary for each Title.
# Add Years as a column in your pivot table.
# Hint: Use Pandas doc(see References in README.md) to explore the pivot_table() arguments
```

```python
employee.pivot_table(index= ['Title'], values = ['Years','Salary'], aggfunc = {'Years':'mean','Salary':'mean'})
```

```python
# We want to know the information about the mean salary and number of employees in each department and for each job title.
```

```python
employee.pivot_table(index= ['Department','Title'], values = ['Salary','Name'], aggfunc = {'Salary':'mean','Name':'count'})
```

```python
# We want to know for each department and for each job title the Salary's median, minimum, maximum and standard deviation
```

```python
employee.pivot_table(index= ['Department','Title'], values = ['Salary'], aggfunc = {'Salary':'describe'}).fillna(0)
```
