```python
# 2.Add an additional column day_type with values 'weekend' and 'workday'
```

```python
# longer queries:
query = "SELECT *,  case  when dayofweek(rental_date) <=5 then 'workday'\
else 'weekend'\
end as 'week flag'\
FROM sakila.rental"
data = pd.read_sql_query(query, engine)
data.head()
```

```python
# 4.Get 10 the longest films.
```

```python
# longer queries:
query = 'SELECT title, length FROM sakila.film  order by length desc limit 10 '


data = pd.read_sql_query(query, engine)
data.head()

```

```python
# -- 5.How many films include Behind the Scenes content?
```

```python
query = 'SELECT count(*) FROM sakila.film  where special_features like "%%Behind the Scenes%%";'


data = pd.read_sql_query(query, engine)
data.head()

```

```python
# -- 6.Which kind of movies (rating) have a mean duration of more than two hours?
select rating, round(avg(length),2) as avg_duration from sakila.film
group by rating
having avg_duration>120
order by avg_duration desc;
```

```python
query = 'SELECT rating,round(avg(length),2) as "avg_duration"  from sakila.film \
group by rating  having avg_duration>120 order by avg_duration desc'


data = pd.read_sql_query(query, engine)
data.head()
```
